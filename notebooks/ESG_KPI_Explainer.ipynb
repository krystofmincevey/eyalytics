## KPI 1: Measures for positive impact on climate change in own operations

- **Information that could also be provided for context on what positive actions cover**:
    * **Energy Consumption**: The total amount of energy used by an individual, community, or organization for various activities. _(Key words: energy consumption, carbon footprint, energy audit, energy management, kilowatt-hour (kWh), energy monitoring, energy use)_
    * **Renewable Energy**: Energy generated from sustainable sources like wind, solar, and hydro, which can be replenished naturally _(Key words: Solar Power, solar panels, Wind Energy, wind turbines, Hydroelectricity, Biomass, Geothermal, Tidal Energy, grid, Net Metering)_
    * **Energy Efficiency**: The practice of using less energy to perform the same task, thereby reducing energy waste _(key words: LED, LED Lighting, HVAC, Insulation, Energy Retrofit, Cogeneration, Building Automation, Smart Grid)_
    * **Energy Certificates**: Documents proving that a certain amount of electricity was generated from renewable or low-carbon sources _(Key words: Renewable Energy Credits (RECs), Carbon Offsets, Green Certificates, Power Purchase agreements (PPA), Energy Labels, Guarantees of Origin, Emission Reduction Certificates, Tradable Renewable Certificates, Carbon Credits, Voluntary Emissions Reduction (VER))_

- **Prompts**:
    * _"Does the report provide information on the company's initiatives or measures related to reducing its carbon footprint and having a positive impact on climate change? Specifically, within the company's direct operations, look for details on:"_
        - **Energy Consumption**:
            - Total kilowatt-hours (kWh) used
            - Energy intensity measures
            - Load profile and peak demand data
            - Results of any energy audits
        - **Renewable Energy**:
            - Types of renewable energy sources used (e.g., solar, wind, hydro)
            - Amount of energy generated or purchased from renewable sources
            - References to grid parity or distributed generation
        - **Energy Efficiency**:
            - Technologies implemented for efficiency (e.g., LED lighting, HVAC efficiency)
            - Measures for building insulation or energy retrofitting
            - Any demand-side management strategies
        - **Energy Certificates**:
            - Types of certificates acquired (e.g., Renewable Energy Credits, carbon offsets)
            - The volume or number of certificates
            - Mention of voluntary or mandatory compliance
        - **Quantifiable Metrics**:
            - Achievements or milestones reached in any of the above areas
            - Year-over-year comparisons or baselines
    * _"What specific actions has the company taken to reduce its energy consumption in direct operations? Are there achievements listed in the report?"_
    * _"Does the report indicate the company's investment in or usage of renewable energy sources, such as wind, solar, or hydro, for its direct operations?"_
    * _"Is there information about the company's energy efficiency improvements? This may include initiatives like energy-efficient lighting, HVAC systems, or industrial processes."_
    * _"Does the report mention the company's acquisition or generation of energy certificates to validate its use of renewable or low-carbon energy sources?"_
    * _"Are there case studies or examples in the report that showcase the company's successful efforts to have a positive impact on climate change through its direct operations?"_

## KPI 2: Career development

- **Information that could also be provided for context**:
    - **Career Development**: The ongoing process of gaining skills, knowledge, and experiences, relevant at all ages and stages, to advance professionally in one's chosen field of work. 
        *Key words*: Skill Development, Job Advancement, Career Path, Mentorship, Professional Development, Career Planning, Job Rotation, Coaching, Employee Retention, Performance Review, Job Satisfaction, Career Milestones, Leadership Development, Continuing Education, Career Assessment, Soft Skills, Hard Skills, Career Change, Career Progression, Employee Training, Job Promotion, Career Counseling, Transferable Skills, Talent Management, Professional Certification, Employee Engagement.

- **Prompts**:
    1. Does the report provide information on the company's initiatives or programs aimed at employee career development? Specifically, search for details on:
        - **Skill Development**:
            - Types of hard and soft skills training offered
            - Inclusion of professional certification programs
        - **Job Advancement & Career Path**:
            - Presence of a defined career progression ladder
            - Opportunities for job rotation or promotion
        - **Mentorship & Coaching**:
            - Availability of formal or informal mentorship programs
            - Employee coaching and leadership development initiatives
        - **Professional Development**:
            - Continuing education benefits or subsidies
            - Employee engagement in career planning or career assessment tools
        - **Performance & Retention**:
            - Frequency and type of performance reviews
            - Mention of employee retention rates or strategies
        - **Employee Satisfaction & Engagement**:
            - Measures of job satisfaction or employee engagement surveys
            - Details on career milestones and achievements recognition
        - **Talent Management**:
            - Programs for talent management or talent pipeline development
        - **Quantifiable Metrics**:
            - Evidence of the effectiveness of career development programs, such as increased job promotions, higher employee retention, or improved performance reviews.

            - Initiatives or partnerships aimed at ethical and sustainable sourcing
        - **Sustainable Procurement**:

        - **Risk Assessment**:

        - **Engagement & Partnerships**:
            - Programs or strategies aimed at engaging suppliers for sustainability
        - **Quantifiable Metrics**:
            - Any specific goals, targets, or KPIs related to supplier sustainability
            - Evidence or data demonstrating effectiveness or improvements.

## KPI 5: GHG emissions

- **Information that could also be provided for context**:
    - **Scope 1 Emissions**: Direct emissions from owned or controlled sources. 
        *Keywords*: Direct Emissions, Fuel Combustion, Company Vehicles, On-site Energy Production, Fugitive Emissions.
    - **Scope 2 Emissions**: Indirect emissions from the generation of purchased energy. 
        *Keywords*: Indirect Emissions, Purchased Electricity, Purchased Heat, Grid Emissions Factor, Renewable Energy Credits (RECs).
    - **Scope 3 Emissions**: All other indirect emissions occurring in the value chain. 
        *Keywords*: Supply Chain Emissions, Upstream Emissions, Downstream Emissions, Business Travel, Employee Commuting, Product Lifecycle, Waste Generated, Outsourced Activities, Transportation and Distribution, Capital Goods, Customer Use-Phase Emissions.
    - *Other Keywords*: Greenhouse Gas (GHG) Emissions, Carbon Accounting, Carbon Footprint, Emission Inventory, Carbon Reporting, GHG protocol.

- **Prompt**:
    1. "Does the report provide an exhaustive breakdown of the company's greenhouse gas (GHG) emissions across Scope 1, Scope 2, and Scope 3 categories? Specifically, look for:
        - **Scope 1 Emissions**:
            - Direct Emissions: Amount and types of direct emissions from company operations
            - Fuel Combustion: Details on fuel types used in company vehicles or on-site energy production
            - Fugitive Emissions: Any emissions from leaks or other unintentional releases
        - **Scope 2 Emissions**:
            - Indirect Emissions: Volume of emissions from purchased electricity or heat
            - Grid Emissions Factor: Metrics used to calculate emissions from purchased energy
            - Renewable Energy Credits (RECs): Whether RECs are used to offset emissions
        - **Scope 3 Emissions**:
            - Supply Chain Emissions: Information on emissions from upstream and downstream activities
            - Business Travel & Employee Commuting: Emissions attributed to these activities
            - Product Lifecycle: Emissions from the entire lifecycle of products or services
            - Outsourced Activities: Emissions from subcontracted or outsourced operations
        - **Other Aspects**:
            - Carbon Accounting: Methodology used for tracking and reporting emissions
            - Emission Inventory: Comprehensive list or database of all emission sources
            - Carbon Reporting: Frequency and transparency of emission reports
            - GHG Protocol: Adherence to global or industry-specific GHG accounting standards

## KPI 6: Employee turnover rate

- **Information that could also be provided for context**:
    - **Employee Turnover Rate**: The percentage of employees who leave an organization during a specified period, often calculated annually.
        *Keywords*: Employee Retention, Attrition Rate, turnover, voluntary / involuntary turnover, turnover rate, Talent Management, Human Resources.

- **Prompt**:
    1. "Does the report provide comprehensive data and insights on the company's employee turnover rate? Specifically, look for:
        - **Annual Turnover/Attrition Rate**: The percentage of employees who left during the last fiscal year
        - **Voluntary vs. Involuntary Turnover**: Differentiation between employees leaving by choice and those leaving due to company decisions
        - **Quantifiable Metrics**: Any specific KPIs or goals related to turnover

## KPI 7: Non-discrimination policy

- **Information that could also be provided for context**:
    - **Non-Discrimination Policy**: A formal document, overseen by management, that outlines an organization's commitment to treat all employees, applicants, and stakeholders equally.
        *Keywords*: race, gender, age, disability, religion, social background, ethnicity, marital status, trade union association, military status, sexual orientation, Equal Opportunity, non-discrimination, diversity, diversity and inclusion (D&I), Human Resources, Inclusion, Gender Equality, Code of Conduct, Employee Handbook, Diversity policy, Non-discrimination policy.

- **Prompt**:
    1. "Does the report discuss the organization's Non-Discrimination Policy in detail? Specifically, look for:
        - **Name of the policy**: Whether the policy is part of these documents (Code of Conduct/Employee Handbook) or part of a separate document (Non-discrimination policy, diversity policy)
        - **Commitment**: Statements affirming commitment to equal opportunities for all
        - **Policy Coverage**: Types of discrimination the policy aims to prevent
        - **Governance**:
            - Role of employees/departments in implementing, monitoring, and enforcing the policy
            - Role of management in implementing, monitoring, and enforcing the policy
